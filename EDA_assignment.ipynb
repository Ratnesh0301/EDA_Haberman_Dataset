```python
# PDF of age, nodes,  operation year,status

sns.FacetGrid(df, hue='status', height=7) \
   .map(sns.distplot, "age")
plt.xlabel("Age")
plt.ylabel("Probability of RV X at age exactly equal to x")
plt.legend(('Survived more than 5 years','Died within 5 years'))
plt.title("Plot 1: PDF of Age")
plt.show()


sns.FacetGrid(df, hue='status', height=7) \
   .map(sns.distplot, "nodes")
plt.xlabel("Nodes")
plt.ylabel("Probability of RV X at nodes exactly equal to x")
plt.legend(('Survived more than 5 years','Died within 5 years'))
plt.title("Plot 2: PDF of Nodes")
plt.show()


sns.FacetGrid(df, hue='status', height=7) \
   .map(sns.distplot, "year")
plt.xlabel("Year")
plt.ylabel("Probability of RV X at operation year exactly equal to x")
plt.legend(('Survived more than 5 years','Died within 5 years'))
plt.title("Plot 3:PDF of Operation Year")
plt.show()
```

### Observations:

PDF(Probability Density Function) shows the density of the feature or number of data present on that point. PDF will be a peak like structure represents high peak if more number of data is present and small peak or no peak is number of data present is less. 

1. Plot 1(PDF of Age) shows that the age range from 30-75 the number of data points present are same or we can day that the both follows the same pattern. we cannot say more from the PDF of Age

2. Plot 2(PDF of Nodes) shows that the if there are less number of nodes then the survival chances are higher than the patients with high number of nodes. The if else statments can be used to distinguish between the patients with status 1 or status 2

 * if number of nodes is equal to zero then patient can survive more than 5 years, or if number of nodes is greater than zero but leas than 3, then patient is having good chances of long survival. But if the number of nodes is greater than 3 then patient died within 5 years.

3. Plot 3(PDF of Operation Year) We cannot predict or conclude anything from the pdf of operation year as the peak is same for all the data points. 


```python
# CDF of age, nodes,  operation year,status

status_1 = df.loc[df['status'] == 1]
status_2 = df.loc[df['status'] == 2]

#CDF of Age
counts, bin_edges = np.histogram(status_1['age'],bins=20,density=True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF(survived)')
plt.plot(bin_edges[1:],cdf,label='CDF(survived)')

counts, bin_edges = np.histogram(status_2['age'],bins=20,density=True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF(Died within 5 years)')
plt.plot(bin_edges[1:],cdf,label='CDF(Died within 5 years)')
plt.xlabel("Age")
plt.ylabel("Probability of RV X at age less than or equal to x")
plt.title("CDF of Age")
plt.legend()
plt.show()
```

### CDF of AGE 

The CDF of Age for patients with survival status 1 and survival status 2 are similiar to each other. There is no prediction made from the CDF of age.

```python
#CDF of Nodes
counts, bin_edges = np.histogram(status_1['nodes'],bins=10,density=True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF(survived)')
plt.plot(bin_edges[1:],cdf,label='CDF(survived)')

counts, bin_edges = np.histogram(status_2['nodes'],bins=10,density=True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF(Died within 5 years)')
plt.plot(bin_edges[1:],cdf,label='CDF(Died within 5 years)')
plt.xlabel("Nodes")
plt.ylabel("Probability of RV X at nodes less than or equal to x")
plt.title('CDF of Nodes')
plt.legend()
plt.show()
```

### CDF of Nodes

From the CDF of survial status 1(orange line in the above plot) which is cumulative sum of PDFs shows that if the number of nodes are less than 5 then there is around 85 % people have good chances of survival. From the red line we can see that if the number of nodes are greater than 40 then there is 100 % people have less chances of survival.

```python
#CDF of Year
counts, bin_edges = np.histogram(status_1['year'],bins=10,density=True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF(survived)')
plt.plot(bin_edges[1:],cdf,label='CDF(survived)')

counts, bin_edges = np.histogram(status_2['year'],bins=10,density=True)

pdf = counts/sum(counts)
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF(Died within 5 years)')
plt.plot(bin_edges[1:],cdf,label='CDF(Died within 5 years)')
plt.xlabel("Year")
plt.ylabel("Probability of RV X at operation year less than or equal to x")
plt.title("CDF of Year")
plt.legend()
plt.show()
```

### CDF of Year

As the CDF of year is similiar for both the survival status 1 and survival status 2. There is no prediction can be made

### Mean, Variance and Standard Deviation

```python
print("\nMean of Nodes:")
print("Long Survived:",np.mean(status_1['nodes']))
print("Short Survived:",np.mean(status_2['nodes']))

print("\nSTD of Nodes:")
print("Long Survived:",np.std(status_1['nodes']))
print("Short Survived:",np.std(status_2['nodes']))
```

### Observations:

1. Mean for long survival is 2.79 and mean of short survival is 7.52 which is much higher than mean of long survival.
2. Standard deviation for long survival is 5.85 and standard deviation for short survival is 9.16 indicates that the spread of nodes for short survival is higher.

### Median, Quantiles, Percentiles and IQR


```python
print('\nMedians:')
print("Long Survived:",np.median(status_1['nodes']))
print("Short Survived:",np.median(status_2['nodes']))

print('\nQunatiles:')
print("Long Survived:",np.percentile(status_1['nodes'],np.arange(0,100,25)))
print("Short Survived:",np.percentile(status_2['nodes'],np.arange(0,100,25)))

print('\n90th Percentile:')
print("Long Survived",np.percentile(status_1['nodes'],90))
print("Short Survived",np.percentile(status_2['nodes'],90))

from statsmodels import robust
print('\nMedian Absolute Deviation')
print(robust.mad(status_1['nodes']))
print(robust.mad(status_2['nodes']))
```

### Observations:

1. The median is the center of the data. the average long survival patient have 0 nodes while the average short survival patients have 4 nodes.
2. From the quantiles for long survival we can see that around 50% of nodes are 0.
3. From the quantiles for short survival, for 75% of patients have minimum 11 nodes detected.
4. At 90th percentile, if node is gretaer than 8 then it has a long survival status and if nodes are greater than 20 then it has a short survival status.
5. Median Absolute Deviation is the simple way to quantify variation. The MAD for nodes with long survival status is 0 and MAD for nodes with short survival status is around 5.93.
